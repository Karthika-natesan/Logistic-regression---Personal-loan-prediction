### 8. Store the target column (i.e.Personal Loan) in the y variable and the rest of the columns in the X variable.


```python
x = df.drop('Personal_Loan',axis=1)
y = df['Personal_Loan'] 
```

### 9. Split the dataset into two parts (i.e. 70% train and 30% test). and standardize the columns using the z-score scaling approach.


```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
scalar = StandardScaler()
```

```python
x_train = scalar.fit_transform(x_train)
x_test = scalar.fit_transform(x_test)
```

### 10. Train and test a Logistic Regression model to predict the likelihood of a liability customer buying personal loans. Display the train and test accuracy scores.

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
loan_prediction = lr.predict(x_test)
```

```python
score = lr.score(x_test,y_test)
```

```python
score
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,loan_prediction))
```

```python
print(accuracy_score(y_test,loan_prediction))
```

### 11. Print the confusion matrix and classification report for the model and write your conclusions on the results

```python
print(confusion_matrix(y_test,loan_prediction))
```

```python
cm = metrics.confusion_matrix(y_test, loan_prediction)
plt.clf()
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Wistia)
classNames = ['no', 'yes']
plt.title('Confusion Matrix - Test Data')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)
s = [['G1', 'G2'], ['G1','G2']]
 
for i in range(2):
    for j in range(2):
        plt.text(j,i, str(s[i][j])+" = "+str(cm[i][j]))
plt.show()
```

* From the above results we can understand this model has predicted true positive values : 1333 out of 1351 (did not buy loan) and true negative values : 88 out of 149 (applied loan)
